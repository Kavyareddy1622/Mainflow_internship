```python

import random
import re
import string

KNOWLEDGE_BASE = [
{"tag":"greeting","patterns":["hello","hi","hey","good morning","good evening","greetings"],"responses":["Hello! Welcome to CollegeBot.","Hi there! I am CollegeBot.","Hey! Great to see you."]},
{"tag":"farewell","patterns":["bye","goodbye","see you","take care","quit","exit"],"responses":["Goodbye!","Take care!","See you later!"]},
{"tag":"admission","patterns":["how to apply","admission process","enrollment process","registration process"],"responses":["Visit the college website and fill the application form."]},
{"tag":"fees","patterns":["fees","fee structure","tuition fee","semester fees"],"responses":["Engineering: Rs 80,000/year. Commerce: Rs 40,000/year."]},
{"tag":"courses","patterns":["courses","degrees","available programs","departments"],"responses":["We offer B.Tech, BCA, BBA, MBA and MCA."]},
{"tag":"hostel","patterns":["hostel","accommodation","dormitory"],"responses":["Hostel facilities are available for boys and girls."]},
{"tag":"placements","patterns":["placement","salary","job","package"],"responses":["Highest package Rs 18 LPA. Average package Rs 6.5 LPA."]},
{"tag":"thanks","patterns":["thanks","thank you","awesome","great"],"responses":["You are welcome!","Happy to help!"]},
{"tag":"unknown","patterns":[],"responses":["I do not understand. Please ask about admissions, fees, courses, hostel or placements."]}
]

def preprocess(text):
    text=text.lower()
    text=text.translate(str.maketrans('','',string.punctuation))
    text=' '.join(text.split())
    return text

def find_best_match(user_input):
    cleaned=preprocess(user_input)
    user_words=set(cleaned.split())
    best_tag='unknown'
    best_score=0
    for entry in KNOWLEDGE_BASE:
        if entry['tag']=='unknown':
            continue
        for pattern in entry['patterns']:
            overlap=len(user_words & set(preprocess(pattern).split()))
            if overlap>best_score:
                best_score=overlap
                best_tag=entry['tag']
    for entry in KNOWLEDGE_BASE:
        if entry['tag']==best_tag:
            return best_tag, random.choice(entry['responses'])

REGEX_PATTERNS=[
(r'\b(hi|hello|hey|greetings)\b','greeting'),
(r'\b(bye|goodbye|exit|quit)\b','farewell'),
(r'\b(fee|fees|tuition|cost)\b','fees'),
(r'\b(admission|apply|join)\b','admission'),
(r'\b(course|courses|degree)\b','courses'),
(r'\b(hostel|accommodation)\b','hostel'),
(r'\b(job|salary|placement)\b','placements'),
(r'\b(thanks|thank)\b','thanks')
]

def regex_match(user_input):
    cleaned=preprocess(user_input)
    for pattern,tag in REGEX_PATTERNS:
        if re.search(pattern,cleaned):
            for entry in KNOWLEDGE_BASE:
                if entry['tag']==tag:
                    return tag, random.choice(entry['responses'])
    return find_best_match(user_input)

def chat():
    print("CollegeBot")
    while True:
        user_input=input("You: ")
        tag,response=regex_match(user_input)
        print("Bot:",response)
        if tag=="farewell":
            break

if __name__=="__main__":
    chat()
```
