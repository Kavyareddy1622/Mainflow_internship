```python
# ==========================================================
# TASK 2: Custom CNN with Data Augmentation & CIFAR-10
# Complete Single-File Implementation
# ==========================================================

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    BatchNormalization,
    GlobalAveragePooling2D,
    Dense,
    Dropout,
    Input
) # Added Input layer
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ==========================================================
# 1. LOAD CIFAR-10 DATASET
# ==========================================================

(x_train, y_train), (x_test, y_test) = cifar10.load_data()

print("Training Images Shape :", x_train.shape)
print("Testing Images Shape  :", x_test.shape)

# ==========================================================
# 2. CLASS NAMES
# ==========================================================

class_names = [
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'
]

# ==========================================================
# 3. VISUALIZE ALL 10 CLASSES
# ==========================================================

plt.figure(figsize=(12, 5))

for i in range(10):
    idx = np.where(y_train == i)[0][0]

    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[idx])
    plt.title(class_names[i])
    plt.axis('off')

plt.suptitle("CIFAR-10 Sample Images")
plt.tight_layout()
plt.show()

# ==========================================================
# 4. PREPROCESSING
# ==========================================================

# Normalize images to [0,1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# One-hot encoding
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("\nData Normalized")
print("Labels One-Hot Encoded")

# ==========================================================
# 5. DATA AUGMENTATION (5 TRANSFORMATIONS)
# ==========================================================

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.20,
    horizontal_flip=True
)

datagen.fit(x_train)

# ==========================================================
# 6. VISUALIZE AUGMENTED IMAGES
# ==========================================================

sample = np.expand_dims(x_train[0], axis=0)

aug_iter = datagen.flow(sample, batch_size=1)

plt.figure(figsize=(8, 8))

for i in range(9):
    batch = next(aug_iter)

    plt.subplot(3, 3, i + 1)
    plt.imshow(batch[0])
    plt.axis('off')

plt.suptitle("Augmented Images")
plt.tight_layout()
plt.show()

# ==========================================================
# 7. DEEP CNN MODEL
# ==========================================================

model = Sequential([
    Input(shape=(32,32,3)), # Added Input layer here
    # Block 1
    Conv2D(
        32,
        (3,3),
        activation='relu',
        padding='same'
    ),
    BatchNormalization(),

    Conv2D(
        32,
        (3,3),
        activation='relu'
    ),

    MaxPooling2D((2,2)),
    Dropout(0.25),

    # Block 2
    Conv2D(
        64,
        (3,3),
        activation='relu',
        padding='same'
    ),
    BatchNormalization(),

    Conv2D(
        64,
        (3,3),
        activation='relu'
    ),

    MaxPooling2D((2,2)),
    Dropout(0.25),

    # Block 3
    Conv2D(
        128,
        (3,3),
        activation='relu',
        padding='same'
    ),
    BatchNormalization(),

    Conv2D(
        128,
        (3,3),
        activation='relu'
    ),

    MaxPooling2D((2,2)),
    Dropout(0.25),

    # Global Average Pooling
    GlobalAveragePooling2D(),

    Dense(128, activation='relu'),
    Dropout(0.3),

    Dense(10, activation='softmax')
])

# ==========================================================
# 8. COMPILE MODEL
# ==========================================================

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nMODEL SUMMARY")
model.summary()

# ==========================================================
# 9. CALLBACKS
# ==========================================================

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    verbose=1
)

# ==========================================================
# 10. TRAIN MODEL
# ==========================================================

history = model.fit(
    datagen.flow(
        x_train,
        y_train_cat,
        batch_size=64
    ),
    epochs=40,
    validation_data=(x_test, y_test_cat),
    callbacks=[early_stop, reduce_lr]
)

# ==========================================================
# 11. PLOT TRAINING HISTORY
# ==========================================================

plt.figure(figsize=(12,5))

# Accuracy
plt.subplot(1,2,1)

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(['Train','Validation'])

# Loss
plt.subplot(1,2,2)

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(['Train','Validation'])

plt.tight_layout()
plt.show()

# ==========================================================
# 12. EVALUATE MODEL
# ==========================================================

loss, accuracy = model.evaluate(
    x_test,
    y_test_cat,
    verbose=0
)

print("\n==============================")
print(f"Test Accuracy : {accuracy*100:.2f}%")
print("==============================")

# ==========================================================
# 13. SAVE MODEL
# ==========================================================

model.save("cifar10_cnn_model.h5")

print("\nModel Saved Successfully")
print("File Name: cifar10_cnn_model.h5")

# ==========================================================
# 14. PREDICT ON 10 TEST IMAGES
# ==========================================================

predictions = model.predict(x_test[:10])

plt.figure(figsize=(15,6))

for i in range(10):

    predicted_class = np.argmax(predictions[i])
    confidence = np.max(predictions[i]) * 100

    plt.subplot(2,5,i+1)

    plt.imshow(x_test[i])

    plt.title(
        f"Pred: {class_names[predicted_class]}\n"
        f"Conf: {confidence:.1f}%"
    )

    plt.axis('off')

plt.suptitle("Predictions on Test Images")
plt.tight_layout()
plt.show()

# ==========================================================
# 15. VISUALIZE FILTER ACTIVATIONS
# ==========================================================

first_conv_layer = model.layers[1] # Changed index from 0 to 1 as Input layer is now at index 0

activation_model = Model(
    inputs=model.input,
    outputs=first_conv_layer.output
)

sample_image = np.expand_dims(
    x_test[0],
    axis=0
)

activations = activation_model.predict(sample_image)

print("\nActivation Shape:", activations.shape)

# ==========================================================
# 16. DISPLAY FIRST 16 FILTERS
# ==========================================================

plt.figure(figsize=(10,10))

for i in range(16):

    plt.subplot(4,4,i+1)

    plt.imshow(
        activations[0, :, :, i],
        cmap='viridis'
    )

    plt.title(f"Filter {i+1}")
    plt.axis('off')

plt.suptitle("First Conv Layer Activations")
plt.tight_layout()
plt.show()

# ==========================================================
# TASK COMPLETED
# ==========================================================

print("\n✅ CIFAR-10 Loaded")
print("✅ All 10 Classes Visualized")
print("✅ Images Normalized")
print("✅ Labels One-Hot Encoded")
print("✅ Data Augmentation Applied")
print("✅ Deep CNN Built")
print("✅ EarlyStopping Used")
print("✅ ReduceLROnPlateau Used")
print("✅ Accuracy Evaluated")
print("✅ Training History Plotted")
print("✅ Filter Activations Visualized")
print("✅ Model Saved (.h5)")
print("✅ Predictions with Confidence Scores Displayed")
```
